```python
def split(X, splits, max_train_size=None):
#     X, y, groups = indexable(X, y, groups)
    n_samples = X.shape[0]
    n_splits = splits
    n_folds = n_splits + 1
    if n_folds > n_samples:
        raise ValueError(
            ("Cannot have number of folds ={0} greater"
             " than the number of samples: {1}.").format(n_folds,
                                                         n_samples))
    indices = np.arange(n_samples)
    print(indices)
    test_size = (n_samples // n_folds)
    print(n_samples)
    print(n_folds)
    print(n_samples % n_folds)
    test_starts = range(test_size*2 + n_samples % n_folds,
                        n_samples, round(test_size/1.5))
    for test_start in test_starts:
        if max_train_size and max_train_size < test_start:
            pass
#             yield (indices[test_start*2 - max_train_size:test_start*2],
#                    indices[test_start*2:test_start + test_size])
        else:
            yield (indices[:test_start],
                   indices[test_start:test_start + round(test_size/1.5)])
```
