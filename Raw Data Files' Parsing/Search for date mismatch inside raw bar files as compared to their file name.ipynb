## Pickle .txt files and compress the pickle

### For 1 security

```python
Files_folder = r"L:\Raw_1_sec_Bar_Data\FX\EURGBP"
file_name = Files_folder.split("\\")[-1]

Indv_Folder = Files_folder+"\\"+"Extracted"
df_for_pickle = pd.DataFrame()
    #print(Indv_Folder)
for file in os.listdir(Indv_Folder):
    df = pd.read_csv(Indv_Folder+"\\"+file,delimiter="|", header=0,index_col=0,parse_dates=True)
    df_for_pickle = pd.concat([df_for_pickle, df])
pickle_file = open(Files_folder+"\\Pickle\\"+file_name+".pkl","wb")
pickle.dump(df_for_pickle, pickle_file)
pickle_file.close()
file_in = open(Files_folder+"\\Pickle\\"+file_name+".pkl","rb")
file_out = gzip.open(Files_folder+"\\Pickle\\"+file_name+".pkl.gz","wb")
shutil.copyfileobj(file_in,file_out)
file_in.close()
file_out.close()
os.remove(Files_folder+"\\Pickle\\"+file_name+".pkl")
```
