````python
pd.read_sql("""
      SELECT
          country,
          COUNT(name)
      FROM 
         airports
      WHERE
         country='Morocco'
      """,con=db).head()
````

## Airlines

#### 7) Select the last 5 airlines alphabetically

````python
pd.read_sql("""
    SELECT
        *
    FROM
        airlines
    """, con=db).head(5)
````

````python
pd.read_sql("""
    SELECT
        name
    FROM
        airlines
    ORDER BY
        name asc
    LIMIT
        5
    """, con=db)
````

#### 8) Which country has the most *inactive* airlines?

````python
pd.read_sql("""
    SELECT
        country,
        COUNT(active='N')
    FROM
        airlines
    GROUP BY country
    ORDER BY 
        COUNT(active='N') desc
        LIMIT
            1
    """, con=db)
````

#### 9) Which countries have *exactly* 9 active airlines?

Hint: look at the [HAVING clause](https://www.w3schools.com/sql/sql_having.asp)

````python
pd.read_sql("""
    SELECT
        country,
        COUNT(active='Y')
    FROM
        airlines
    GROUP BY country
        HAVING COUNT(active='Y')=9
    
    """, con=db)
````

#### 10) Which airlines contain the word 'green' somewhere in their name?

Hint: look at the [LIKE](https://www.w3schools.com/sql/sql_like.asp) keyword to see how to search in text columns

````python
pd.read_sql("""
    SELECT
        name        
    FROM
        airlines
    WHERE name LIKE '%green%'
    """, con=db)
````

#### 12) Select the first 10 routes as well as the *country of the origin airport*

````python
pd.read_sql("""
    SELECT
        routes.[index],
        routes.source,
        airports.name AS source_airport
    FROM
        routes
        JOIN airports ON routes.source=airports.code
    LIMIT
        10
    """, con=db)
````

#### 15) How many routes are there *into* Brazil?

````python
pd.read_sql("""
    SELECT
        COUNT(routes.[index]),
        routes.dest_id,
        airports.country
    FROM
        routes
        JOIN airports ON routes.dest_id=airports.id
    WHERE airports.country='Brazil'
    """, con=db)
````
