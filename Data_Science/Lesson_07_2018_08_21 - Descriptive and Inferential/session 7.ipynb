```python
import matplotlib.pyplot as plt

from scipy.stats import skewnorm

%matplotlib inline

#4 is a paramenter we use to skew the list, 5000 is the size of the list
r = skewnorm.rvs(4, size=5000)

plt.hist(r, bins=30)
plt.show()
print(len(r))
```

```python
sample_means=[]
np.random.seed(42)

for i in range(1000):
    sample=np.random.choice(r,100)
    sample_means.append(np.mean(sample))
    
plt.hist(sample_means)
plt.show
print(len(sample_means))
```

```python
print(stats.skew(sample_means))
print(stats.skew(r))
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame()

np.random.seed(42)

df["jockeys"] = np.random.normal(150, 10, 100)
df["jockeys_2"] = np.random.normal(150, 10, 100)
df["basketball_players"] = np.random.normal(190, 10, 100)
```

```python
#most important types of graphs
#line charts
#histograms (distributions)
#scatter graphs
```

```python
#10 is thie interval between min and max set of data
plt.hist(df['jockeys'],20)
plt.xlabel('height,cm')
plt.ylabel('number of samples')
plt.title('height distribution of jockeys')
plt.grid()
plt.show()
```

```python
plt.scatter(df['jockeys'],df['basketball_players'])
plt.grid()
plt.xlabel('jockeys height (cm)')
plt.ylabel('basketball players height (cm)')
plt.title('scatter plot')
plt.show()
```

```python
plt.plot(df['basketball_players'])
plt.grid()
plt.ylabel('height (cm)')
plt.ylim(0,250)
plt.title('Basketball players')
```
