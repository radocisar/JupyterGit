## Data Visualisation With Python

---

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ufo_df = pd.read_csv("datasets/ufo.csv")
drinks = pd.read_csv("datasets/drinks.csv")

# fix North America ambiguity
continent_column = drinks['continent'].fillna(value='NorthAmerica').copy()

drinks_df = drinks.drop('continent', axis=1).copy()
drinks_df['continent'] = continent_column
```

```python
ufo_df.describe()
```

```python
drinks_df.describe()
drinks_df
```

```python
# use a line chart to plot the beer_servings for all countries

plt.plot(drinks_df["beer_servings"])
plt.grid()
plt.show()
```

```python
# now let's plot the wine_servings too

plt.plot(drinks_df["beer_servings"])
plt.plot(drinks_df["wine_servings"])
plt.grid()
plt.show()
```

```python
# ok, now let's get tidy
#figure is a function of matplotlib library
from matplotlib.pyplot import figure

line_up, = plt.plot(drinks_df["beer_servings"], label="beer")
line_down, = plt.plot(drinks_df["wine_servings"], label="wine")
plt.legend(loc=2, handles=[line_up, line_down], fontsize="x-large")
plt.xlabel("countries")
plt.ylabel("(lt.)")
plt.title("alcoholic servings")
plt.grid()
plt.show()

plt.rcParams['figure.figsize'] = [10, 5]
```

```python
# now let's do a scatter plot of beer servings and wine servings
#rgb color picker can be chosen and put on the graph
#alpha is how transparent the data could be

plt.scatter(drinks_df["beer_servings"], drinks_df["wine_servings"], s=90, c='green', alpha=0.9)
plt.xlabel("beer (lt.)")
plt.ylabel("wine (lt.)")
plt.grid()
plt.show()

plt.rcParams['figure.figsize'] = [10, 5]
```

```python
# now let's divide by continent

# Africa
africa_beer = drinks_df[drinks_df['continent'] == 'AF']['beer_servings']
africa_wine = drinks_df[drinks_df['continent'] == 'AF']['wine_servings']
africal_spirit=drinks_df[drinks_df['continent']=='AF']['spirit_servings']
africa_total=drinks_df[drinks_df['continent']=='AF']['total_litres_of_pure_alcohol']
# Asia
asia_beer = drinks_df[drinks_df['continent'] == 'AS']['beer_servings']
asia_wine = drinks_df[drinks_df['continent'] == 'AS']['wine_servings']
asia_spirit=drinks_df[drinks_df['continent']=='AS']['spirit_servings']
asia_total=drinks_df[drinks_df['continent']=='AS']['total_litres_of_pure_alcohol']
# Europe
europe_beer = drinks_df[drinks_df['continent'] == 'EU']['beer_servings']
europe_wine = drinks_df[drinks_df['continent'] == 'EU']['wine_servings']
europe_spirit=drinks_df[drinks_df['continent']=='EU']['spirit_servings']
eurpoe_total=drinks_df[drinks_df['continent']=='EU']['total_litres_of_pure_alcohol']
# North America
north_america_beer = drinks_df[drinks_df['continent'] == 'NorthAmerica']['beer_servings']
north_america_wine = drinks_df[drinks_df['continent'] == 'NorthAmerica']['wine_servings']
north_america_spirit=drinks_df[drinks_df['continent']=='AS']['spirit_servings']
north_america_total=drinks_df[drinks_df['continent']=='AS']['total_litres_of_pure_alcohol']
# Oceania
oceania_beer = drinks_df[drinks_df['continent'] == 'OC']['beer_servings']
oceania_wine = drinks_df[drinks_df['continent'] == 'OC']['wine_servings']
# South America
south_america_beer = drinks_df[drinks_df['continent'] == 'SA']['beer_servings']
south_america_wine = drinks_df[drinks_df['continent'] == 'SA']['wine_servings']
```

```python
plt.scatter(africa_beer, africa_wine, s=90, c='green', alpha=0.9, label="Africa")
plt.scatter(asia_beer, asia_wine, s=90, c='red', alpha=0.9, label="Asia")
plt.scatter(europe_beer, europe_wine, s=90, c='blue', alpha=0.9, label="Europe")
plt.scatter(north_america_beer, north_america_wine, s=90, c='black', alpha=0.9, label="North America")
plt.scatter(oceania_beer, oceania_wine, s=90, c='orange', alpha=0.9, label="Oceania")
plt.scatter(south_america_beer, south_america_wine, s=90, c='cyan', alpha=0.9, label="South America")
plt.xlabel("beer (lt.)")
plt.ylabel("wine (lt.)")
plt.legend(loc=2, fontsize="medium")
#setting the limits - zooming in on a piece of code
plt.xlim(0,100)
plt.ylim(0,30)
plt.grid()
plt.show()

plt.rcParams['figure.figsize'] = [10, 5]
```
