```python
import pandas as pd
IP = pd.read_table(r"C:\Ping\ping_proquest.txt", delimiter="|", header=None, names=["First", "Second"], encoding="UTF-8")
IP_science = pd.read_table(r"C:\Ping\ping_science_direct.txt", delimiter="|", header=None, names=["First", "Second"], encoding="UTF-8")

print(IP.head(10))
print(IP_science.head(10))
```

```python
print(IP["First"].size)
print(IP_science["First"].size)
```

```python
i = list(range(1,IP["First"].size,3))
i_science = list(range(1,IP_science["First"].size,3))
print(i)
print(i_science)
```

```python
IP_new = IP.drop(index=i,columns=["Second"])
IP_new.reset_index(inplace=True, drop=True)
IP_new_science = IP_science.drop(index=i_science,columns=["Second"])
IP_new_science.reset_index(inplace=True, drop=True)

print(IP_new.head())
print(IP_new_science.head())
```

```python
print(IP_new.tail())
print(IP_new_science.tail())
```

```python
list1 = list(range(1,IP_new["First"].size,2))
list1_science = list(range(1,IP_new_science["First"].size,2))
print(list1)
print(list1_science)
```

```python
IPs = IP_new.loc[list1,["First"]]
IPs.reset_index(inplace=True,drop=True)
IPs_science = IP_new_science.loc[list1_science,["First"]]
IPs_science.reset_index(inplace=True,drop=True)

print(IPs.head())
print(IPs.tail())
print(IPs_science.head())
print(IPs_science.tail())
```

```python
Times = IP_new.drop(index=list1)
Times.reset_index(inplace=True,drop=True)
Times_science = IP_new_science.drop(index=list1_science)
Times_science.reset_index(inplace=True,drop=True)

print(Times.head())
print(Times.tail())
print(Times_science.head())
print(Times_science.tail())
```

```python
Final = pd.concat([Times,IPs],axis=1)
Final.columns = ["First","Second"]
Final_uniques = Final.drop_duplicates(subset=["Second"])
Final_science = pd.concat([Times_science,IPs_science],axis=1)
Final_science.columns = ["First","Second"]
Final_science_uniques = Final_science.drop_duplicates(subset=["Second"])

print(Final_uniques)
print(Final_science_uniques)
```

```python
print(pd.pivot_table(Final,index="Second", aggfunc="count"))
print(pd.pivot_table(Final_science,index="Second", aggfunc="count"))
```
