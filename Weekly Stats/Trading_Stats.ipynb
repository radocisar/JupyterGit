```python
import pandas as pd
import datetime as dt
excel_file = pd.read_excel(r'C:\Users\radoC\Desktop\Trading_Review\Trades.xlsx',sheet_name="Sheet1", index_col=0)
excel_file.head()
```

```python
def figure_Monday(full_date):
    return full_date - dt.timedelta(days=full_date.dayofweek)
    
excel_file["temp_Date"] = excel_file.index
excel_file["Week"] = excel_file["temp_Date"].apply(figure_Monday)
excel_file.drop("temp_Date", axis=1, inplace=True)
excel_file = excel_file[['Week', 'Stock', 'Trade Type', 'Result', 'AM/Lunch/PM/All',
       'Chased?/Too early?', 'Comments', 'In play?', 'Day of play',
       'News or technical play?']]
excel_file
```

```python
excel_file.head()
```

```python
#### Stats by Type of Play
## Total
type_of_play_stats = excel_file.groupby("News or technical play?")["Result"].describe().loc[:,["count","mean","min", "max"]]
type_of_play_stats["Result"] = excel_file.groupby("News or technical play?")["Result"].sum()
type_of_play_stats.index.name = None
#type_of_play_stats
## By Week
type_of_play_by_week_stats = excel_file.groupby(["Week", "News or technical play?"])["Result"].describe().loc[:,["count","mean","min", "max"]]
type_of_play_by_week_stats["Result"] = excel_file.groupby(["Week", "News or technical play?"])["Result"].sum()
type_of_play_by_week_stats.index.names = [None,None]
# type_of_play_stats.sort_values(by=["Result"], ascending=False, inplace=True)
#type_of_play_by_week_stats

#### Stats by Trade Type
## Total
trade_type_stats = excel_file.groupby("Trade Type")["Result"].describe().loc[:,["count","mean","min", "max"]]
trade_type_stats["Result"] = excel_file.groupby("Trade Type")["Result"].sum()
trade_type_stats.index.name = None
trade_type_stats.sort_values(by=["Result"], ascending=False, inplace=True)
#trade_type_stats
## By Week
trade_type_by_week_stats = excel_file.groupby(["Week", "Trade Type"])["Result"].describe().loc[:,["count","mean","min", "max"]]
trade_type_by_week_stats["Result"] = excel_file.groupby(["Week", "Trade Type"])["Result"].sum()
trade_type_by_week_stats.index.name = None
#trade_type_by_week_stats.sort_values(by=["Result"], ascending=False, inplace=True)
#trade_type_by_week_stats

#### Stats by Day of Play
## Total
day_of_play_stats = excel_file.groupby("Day of play")["Result"].describe().loc[:,["count","mean","min", "max"]]
day_of_play_stats["Result"] = excel_file.groupby("Day of play")["Result"].sum()
day_of_play_stats.index.name = None
#day_of_play_stats
## By Week
day_of_play_by_week_stats = excel_file.groupby(["Week", "Day of play"])["Result"].describe().loc[:,["count","mean","min", "max"]]
day_of_play_by_week_stats["Result"] = excel_file.groupby(["Week", "Day of play"])["Result"].sum()
day_of_play_by_week_stats.index.name = None
#day_of_play_by_week_stats

#### Stats by In Play
## Total
in_play_stats = excel_file.groupby("In play?")["Result"].describe().loc[:,["count","mean","min", "max"]]
in_play_stats["Result"] = excel_file.groupby("In play?")["Result"].sum()
in_play_stats.index.name = None
#in_play_stats
## By Week
in_play_by_week_stats = excel_file.groupby(["Week", "In play?"])["Result"].describe().loc[:,["count","mean","min", "max"]]
in_play_by_week_stats["Result"] = excel_file.groupby(["Week", "In play?"])["Result"].sum()
in_play_by_week_stats.index.name = None
#in_play_by_week_stats

#### Stats by Time of Day
## Total
time_of_day_stats = excel_file.groupby("AM/Lunch/PM/All")["Result"].describe().loc[:,["count","mean","min", "max"]]
time_of_day_stats["Result"] = excel_file.groupby("AM/Lunch/PM/All")["Result"].sum()
time_of_day_stats.index.name = None
#time_of_day_stats
## By Week
time_of_day_by_week_stats = excel_file.groupby(["Week", "AM/Lunch/PM/All"])["Result"].describe().loc[:,["count","mean","min", "max"]]
time_of_day_by_week_stats["Result"] = excel_file.groupby(["Week", "AM/Lunch/PM/All"])["Result"].sum()
time_of_day_by_week_stats.index.name = None
#time_of_day_by_week_stats

#### Stats by Chased
## Total
chased_stats = excel_file.groupby("Chased?/Too early?")["Result"].describe().loc[:,["count","mean","min", "max"]]
chased_stats["Result"] = excel_file.groupby("Chased?/Too early?")["Result"].sum()
chased_stats.index.name = None
#chased_stats
## By Week
chased_by_week_stats = excel_file.groupby(["Week", "Chased?/Too early?"])["Result"].describe().loc[:,["count","mean","min", "max"]]
chased_by_week_stats["Result"] = excel_file.groupby(["Week", "Chased?/Too early?"])["Result"].sum()
chased_by_week_stats.index.name = None
#chased_by_week_stats
```

# Overall Stats:

```python
print("Overall Stats:")
print("   P&L: {}".format(excel_file["Result"].sum()))
print("   Total Trades: {}".format(excel_file.describe().loc["count",:][0]))
print("   Max Loss: {}".format(excel_file["Result"].describe()["min"]))
print("   Max Profit: {}".format(excel_file["Result"].describe()["max"]))
print(" ")
print("Play Type Stats:")
print(type_of_play_stats)
print(" ")
print("Trade Type Stats:")
print(trade_type_stats)
print(" ")
print("Day of Play Stats:")
print(day_of_play_stats)
print(" ")
print("In Play Stats:")
print(in_play_stats)
print(" ")
print("Time of Day Stats:")
print(time_of_day_stats)
print(" ")
print("Chased?:")
print(chased_stats)
```
