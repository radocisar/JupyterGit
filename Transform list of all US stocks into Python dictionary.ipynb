```python
import pandas as pd

df = pd.read_excel(r"C:\Python TWS API\Python_TWS_API_Historical_Data_Download\Python_TWS_API_Historical_Data_Download\US_Stock_Tickers.xlsx", skiprows=14, sheet_name="Sheet1")
df.head(10)
```

```python
df_less_MLPs_and_LTD = df[~df["SECURITY_TYP"].isin(["MLP","Ltd Part"])]
print(len(df_less_MLPs_and_LTD))
df_less_MLPs_and_LTD.head(10)
```

```python
df_less_MLPs_and_LTD_and_within_valid_price_range = df_less_MLPs_and_LTD[(df_less_MLPs_and_LTD["PX_CLOSE_1D"]>=1) & (df_less_MLPs_and_LTD["PX_CLOSE_1D"]<500)]
print(len(df_less_MLPs_and_LTD_and_within_valid_price_range))
df_less_MLPs_and_LTD_and_within_valid_price_range.head(10)
```

```python
EXCH_list = df_less_MLPs_and_LTD_and_within_valid_price_range["Bloomberg_to_IB_exchange_mapping"].tolist()
TICKER_list = df_less_MLPs_and_LTD_and_within_valid_price_range["ID_EXCH_SYMBOL"].tolist()
print(EXCH_list)
print(TICKER_list)
```

```python
d = {TICKER_list:EXCH_list for TICKER_list,EXCH_list in zip(TICKER_list,EXCH_list)}
d
```
